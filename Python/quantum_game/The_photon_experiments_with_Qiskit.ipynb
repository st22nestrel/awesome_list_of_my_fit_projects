## A quick example

````python
from qiskit import *
from qiskit_aer import *

qc = QuantumCircuit(2,2) # create a circuit with two quantum bits (qubits) and two classical bits

qc.x(0)
qc.h(1)
qc.cx(0,1)
qc.measure(0,0)
qc.measure(1,1)
qc.draw("mpl")
#print(qc.draw()) #local
````

````python
## execute the circuit 1024 times
job = AerSimulator().run(qc,shots=1024)
# get the result
counts = job.result().get_counts(qc)
print(counts)
````

## Apply NOT gate randomly

````python
from qiskit import *
from qiskit_aer import *
from random import randint

n = 6 # number of qubits

qc = QuantumCircuit(n,n) # create a circuit with n qubits and n classical bits



for i in range(n):
  if randint(0,99) % 2 == 1:
    qc.x(i)

qc.barrier()

for i in range(n):
  qc.measure(i,i)

qc.draw("mpl")
````

### Guess the outcome

````python
# execute the circuit 1024 times
job = AerSimulator().run(qc,shots=1024)
# get the result
counts = job.result().get_counts(qc)
print(counts)
````

## Apply Controlled NOT gate randomly

````python
from qiskit import *
from qiskit_aer import *
from random import randint

n = 6 # number of qubits

qc = QuantumCircuit(n,n) # create a circuit with n qubits and n classical bits



for i in range(n):
  if randint(0,99) % 2 == 1:
    qc.x(i)

cnot_number = 0
while cnot_number<4:
  i = randint(0,n-1)
  j = randint(0,n-1)
  if i != j:
    qc.cx(i,j)
    cnot_number = cnot_number + 1

qc.barrier()

for i in range(n):
  qc.measure(i,i)

qc.draw("mpl")
````

### Guess the outcome

````python
# execute the circuit 1024 times
job = AerSimulator().run(qc,shots=1024)
# get the result
counts = job.result().get_counts(qc)
print(counts)
````

## Photon Experiment 2

````python
q = QuantumRegister(1)
c = ClassicalRegister(1)
qc = QuantumCircuit(q,c)

qc.h(0)

qc.measure(q,c)

qc.h(q[0]).c_if(c,0)

qc.measure(q,c)

qc.draw("mpl")
````

### Outcome

````python
# execute the circuit 1024 times
job = AerSimulator().run(qc,shots=1024)
# get the result
counts = job.result().get_counts(qc)
print(counts)
````
